```python
# Creating a set of unique numbers
unique_numbers = {1, 2, 3, 4, 5}
print(unique_numbers)

# Adding an item to a set
unique_numbers.add(6) # we are not using append() here 
print(unique_numbers)

# Sets don't allow duplicates
unique_numbers.add(3)  # This will not add another '3'
print(unique_numbers)
```

```python
# if else ladder

grade = float(input("Write the marks here"))

if grade == 100:
    print("A+ Perfect Score!!")
elif grade > 90:
    print("A+")
elif grade > 70:
    print("B")
elif grade > 55:
    print("C")
elif grade > 45:
    print("D")
elif grade > 30:
    print("E")
else:
    print("FAIL!!!")
```

```python
for i in range(10): # to move 10 times
    print (i) # printing i 10 times

print("\nWhen we use break")
```

```python
# break takes us out of the loop early
for i in range(10): # to move 10 times
    print (i) # printing i 10 times
    if i == 5:
        break
```

```python
print("\nThis happens when continue is used")

# skip the next iteration
for i in range(10): # to move 10 times
    if i == 5:
        continue

    print (i) # printing i 10 times
```

```python
print("\nThis happens when Pass is used")

# skip the next iteration
for i in range(10): # to move 10 times
    if i == 5:
        pass
    print (i) # printing i 10 times
```
