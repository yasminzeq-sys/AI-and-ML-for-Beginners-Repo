    <h4>  How is python code executed?
</h4> </div>

- If we have some C code, we need a complier to convert it into machine code that the computer can understand and execute directly.

- C Code --(C Compiler)--> Machine Code --(Executed by CPU)

Machine code is specific to the hardware architecture of the computer,

- so a C compiler generates machine code tailored to the target platform, 
- e.g x86, ARM, or different hardware of windows(commonly x86 intel or amd) PCs vs Mac(commonly ARM or Apple Silicon).


Just like how different people speak different languages, computers also have their own "languages" called machine code.

Java on the other hand uses a different approach.

- Java Code --(Java Compiler)--> Bytecode --(Java Virtual Machine)--> Machine Code --(Executed by CPU)

Java compiler translates Java code into an intermediate form called **bytecode**, which is platform-independent.

This bytecode can be executed on any platform that has a Java Virtual Machine (JVM), making Java applications highly portable.

The JVM interprets the bytecode and translates it into machine code specific to the underlying hardware, allowing it to be executed by the CPU.

We have Windows JVM, Mac JVM, Linux JVM etc.

C# and Python have similar approaches, so they are platform independent. 

- CPython --> **python Bytecode** --> Python Virtual Machine (PVM) --> Machine Code --> Executed by CPU

 
Now if similarly, we have Jython, 

- Jython Code --(Jython Compiler)--> Java Bytecode --(Java Virtual Machine)--> Machine Code --(Executed by CPU)

Since Jython compiles Python code into Java bytecode, it can run on any platform with a JVM, we can also put any java code inside Jython and use it seamlessly.

### Boolean

- Represensents the truth values, either `True` or `False`.
- These are related to the boolean logic of maths and computer science.

**Boolean Operations**

- **Boolean literals:** `True`, `False`
- **Comparisons** (result is bool): `==  !=  >  <  >=  <=`
- **Logical operators:** `and` (both), `or` (either), `not` (negate)


```python
# -- Basics: True/False and comparisons
print(True, False)                           # output: True False
print(5 == 5, 5 != 3)                        # output: True True
print(7 > 2, 3 >= 3, 4 < 1)                  # output: True True False

# -- Logical ops: and / or / not
print(True and True)                         # output: True
print(True and False)                        # output: False
print(True or False)                         # output: True
print(not False)                             # output: True
```

- **Truthiness (bool(x))**  
  Falsy: `0, 0.0, 0j, "" (empty string), [], (), {}, set(), range(0), None`  
  Everything else is truthy.

```python
# -- Truthiness: using the bool(x) function 
# bool is tricky in python

print(bool(0), bool(""), bool([]))           # output: False False False (all empty/zero)

# Everything else is True
print(bool(42), bool(0.1), bool(-1), bool("Hi"), bool("False"))                  # output: True True
```

One final Trick with python 

- **Short-circuit behavior** (returns the *operand*, not forced to bool): 

  - `A and B` → returns `A` if `A` is falsy, else returns `B`  

  - `A or B`  → returns `A` if `A` is truthy, else returns `B`
  
  (Check this - Common idiom: `name = user_input or "Guest"`)

```python
# -- Short-circuit returns the operand (not forced to bool)

x = "" or "text"

# we have seen that empty string is False
# can we compare a boolean with string? 
# we do that and see the result

print(x)
```

```python
# -- Short-circuit returns the operand (not forced to bool)

user = ""                                    # output: empty string is falsy
greeting_name = user or "Guest"              # output: returns "Guest" because user is falsy
print(greeting_name)                         # output: Guest

```

### Type Conversion

**Type casting**, also known as type conversion, is a process in programming where you convert a variable from one data type to another. 

You can convert between data types using type conversion functions like `int()`, `float()`, `str()`, etc.

Using the function `type()`, we can check the type of any variable or value in Python.

- functions are like little machines that take inputs, do something with them, and give back an output.

- we use them in our code to perform specific tasks without rewriting the same code repeatedly.

- Other functions we have used are `print()`, `input()`, `len()` etc.

```python
# Checking Type

# Integer
x = 10
print("x is of type:", type(x))

# Float
y = 3.14
print("y is of type:", type(y))

# String
name = "John"
print("name is of type:", type(name))

# Boolean
is_active = True
print("is_active is of type:", type(is_active))
```

    <h4>Operators in Python

Operators are used to perform operations on variables and values.

**Arithmetic Operators** 

Common arithmetic operators include: `+`, `-`, `*`, `/`, `//` (floor division), `%` (modulus), and `**` (exponentiation).


1. Division (/): 
    - Always results in a floating-point number.
    
<br>

2. Floor Division (//):
    - Divides and returns the largest integer less than or equal to the result.
    
<br>

3. Modulus (%): 
    - Returns the remainder of the division.
    
<br>


4. Exponentiation (**): 
    - Raises a number to the power of another.
    
<br>

5. Methods:
    - bit_length(): Returns the number of bits necessary to represent the integer in binary.


```python
# Arithmetic operations
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)

print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
```

**Comparison Operators**

Comparison operators include: `==`, `!=`, `>`, `<`, `>=`, `<=`.
These return a boolean value (`True` or `False`).


```python
# Comparison operations
x = 10
y = 20

print("Is x equal to y?", x == y)
print("Is x not equal to y?", x != y)
print("Is x greater than y?", x > y)
print("Is x less than or equal to y?", x <= y)
```

**Logical Operators**

Logical operators are used to combine conditional statements. 
- `and`: Returns True if both statements are true
- `or`: Returns True if one of the statements is true
- `not`: Reverses the result


```python
# Logical operations
x = 5
y = 10

print("x > 0 and y > 0:", x > 0 and y > 0)
print("x > 0 or y < 0:", x > 0 or y < 0)
print("not(x > 0):", not(x > 0))
```

- **Precedence:** `not` > `and` > `or` (use parentheses if unsure)


```python
# -- Precedence: not > and > or

expr1 = True or False and False              # output: evaluates as True or (False and False) -> True or False -> True
expr2 = (True or False) and False            # output: parentheses change it -> True and False -> False

print(expr1, expr2)                          # output: True False
```

- **Chained comparisons:** `a < b < c` (reads naturally, no need to repeat `b`)


```python
# -- Chained comparisons (reads naturally)
age = 21
print(18 <= age < 60)                        # output: True (age between 18 and 59 inclusive of 18)
```

- **Equality vs Identity:**  
  - `==` compares **values**  
  - `is` compares **objects** (use `is None`, not `== None`)

```python
# -- Equality vs identity
a = [1, 2]; b = [1, 2]

print(a == b)                                # output: True  (same values)
print(a is b)                                # output: False (different list objects)


n = None
print(n is None)                             # output: True  (preferred way to check None)
```

```python
# Simple calculator

print("Welcome to the simple calculator!")
print("Choose an operation: +, -, *, /")

# User input for the operation
operation = input("Enter operation: ")

# User input for two numbers
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Perform the chosen operation
if operation == "+":
    result = num1 + num2
elif operation == "-":
    result = num1 - num2
elif operation == "*":
    result = num1 * num2
elif operation == "/":
    result = num1 / num2
else:
    result = "Invalid operation"

print(f"The result is: {result}")
```
