</h4> </div>

**What are conditionals?**  
Conditionals let your program **choose** what to do based on a **True/False** test.

**Core keywords** (Conditional Statements):
- `if` ‚Äî run a block **only if** a condition is True  
- `elif` ‚Äî try another condition **if previous were False**  
- `else` ‚Äî fallback when none of the above matched

**Syntax rules**
- End the condition line with a **colon `:`**
- The following block must be **indented** (4 spaces is standard)
- Conditions are any expressions that return a boolean value **True/False**


````python
# 1) Basic if
age = 20                                     # output: store an integer
if age >= 18:                                 # output: condition is True (20 >= 18)
    print("Adult")                            # output: Adult
````

````python
# 2) if / else
temp_c = 30                                   # output: store temperature
if temp_c > 35:                               # output: condition is False (30 > 35 -> False)
    print("Hot")                              # output: skipped
else:
    print("Comfortable or cool")              # output: Comfortable or cool
````

````python
# 3) if / elif / else
score = 76                                    # output: store a score

if score >= 90:                               # output: False
    print("A")                                # output: skipped
elif score >= 75:                             # output: True (76 >= 75)
    print("B")                                # output: B
else:
    print("C")                                # output: skipped
````

**Grade Calculator using Conditional Statements**

````python
# Grade Calculator

# Input: Get the student's score
score = int(input("Enter the student's score (0-100): "))

# Determine the grade based on the score
if score >= 90:
 grade = "A"
elif score >= 80:
 grade = "B"
elif score >= 70:
 grade = "C"
elif score >= 60:
 grade = "D"
else:
 grade = "F"

# Output: Display the grade
print(f"The student's grade is: {grade}")
````

**Handy helpers**

- **`range(n)`** ‚Üí 0..n-1 numbers for counted loops.
- **`enumerate(seq)`** ‚Üí pairs of (index, item).

````python
# --- C) Summation with a for loop
total = 0                            # output: start from 0
for n in range(1, 6):                # output: n takes 1..5
    total = total + n                # output: accumulate sum
print("sum 1..5 =", total)           # output: sum 1..5 = 15
````

````python
# can also use total +=

prices = [10.0, 12.5, 7.5]                       # output: three prices
total = 0.0                                      # output: start total at 0
for p in prices:                                 # output: loop through each price
    total += p                                   # output: same as total = total + p
print("Total:", total)                           # output: Total: 30.0

print()                                          # output: blank line
````

**Enumerate**
- `enumerate(seq)`: gives pairs of (index, item) from a sequence.

- enumerate is a built-in function that adds a counter (index) to any iterable (like a list, string, tuple) when you loop over it.

It gives you a **pair** for each item: **(index, value)**.  
- Index starts at **0** by default.
- You can start from **1** with `enumerate(things, start=1)`

You get the **position** and the **item** together‚Äîcleaner than writing `for i in range(len(...))`.


````python
# --- D) enumerate: index + value together
name = "Riya"                        # output: a short string
for idx, ch in enumerate(name):      # output: idx=0..3, ch='R','i','y','a'
    print(idx, ch)                   # output: 0 R / 1 i / 2 y / 3 a
````

````python
# enumerate over a list (index + item)
fruits = ["apple", "banana", "cherry"]           # output: a simple list of 3 fruits
for i, item in enumerate(fruits):              # output: pairs: (0,'apple'), (1,'banana'), (2,'cherry')
    print(i, item)                             # output: 0 apple / 1 banana / 2 cherry

print()                                          # output: blank line for readability
````

#### **Helpful controls**

- **`break`**: stop the loop early.
- **`continue`**: skip to the next iteration.
- **`pass`**: do nothing (placeholder).

````python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]

for fruit in fruits:
   print(fruit)
````

````python
# using break to exit loop
for fruit in fruits:
   if fruit == "banana":
       break
   print(fruit)
````

````python
# --- E) break: stop early when a condition is met
found_over_7 = None                  # output: placeholder
for n in range(5, 11):               # output: 5..10
    if n > 7:                        # output: True when n is 8
        found_over_7 = n             # output: remember 8
        break                        # output: stop the loop
print("first > 7:", found_over_7)    # output: first > 7: 8
````

````python
# --- F) continue: skip certain iterations
for n in range(1, 8):                # output: 1..7
    if n % 2 == 0:                   # output: True for even numbers
        continue                     # output: skip printing evens
    print("odd:", n)                 # output: odd: 1 / 3 / 5 / 7
````

````python
# --- G) while loop: repeat while condition is True
count = 3                            # output: start at 3

while count > 0:                     # output: True for 3,2,1 then stops at 0
    print("count:", count)           # output: count: 3 / 2 / 1
    count = count - 1                # output: decrease by 1
````

````python
# --- H) while loop for accumulation until a threshold
s = 0                                # output: sum starts 0
x = 1                                # output: next number to add
while s < 10:                        # output: run until s reaches 10 or more
    s = s + x                        # output: add x to sum
    x = x + 1                        # output: move to next number
print("sum reached:", s)             # output: sum reached: 10 or more
````

**Nested Loops**

- To handle complex data (like tables), you can put one loop **inside** another loop.
- The inner loop runs completely for each iteration of the outer loop.
- This is useful for working with multi-dimensional data structures.
- Like matrices or grids.

````python
matrix = [
    [1, 2, 3],                             # output: row 0
    [4, 5, 6],                             # output: row 1
    [7, 8, 9],                             # output: row 2
]
````

````python
print(len(matrix))                          # output: 3 (number of rows)
print((len(matrix), len(matrix[0])))       # to get shape # output: (3, 3) (rows, columns)
````

````python
# Print elements row by row
for r in range(len(matrix)):               # output: r = 0..2
    for c in range(len(matrix[0])):        # output: c = 0..2
        print(matrix[r][c], end=" ")       # output: prints each element with a space
    print()                                # output: newline per row

print()                                    # output: blank line
````

````python
# Example: Multiplication Table
for i in range(1, 4): # Outer loop
    for j in range(1, 6): # Inner loop
        print(f"{i} x {j} = {i * j}")
print() # Adds a blank line after each table
````

````python
# --- I) Nested loops: tiny 2x3 grid
for r in range(2):                   # output: r=0 then r=1
    for c in range(3):               # output: c=0,1,2
        print(f"({r},{c})", end=" ") # output: (0,0) (0,1) (0,2) then (1,0) (1,1) (1,2)
    print()                          # output: newline after each row
````

**Star Patterns**

````python
n = 5                                      # output: height/size of the pattern

# A) Left-aligned triangle (nested loops)
for row in range(1, n + 1):                # output: row = 1..5
    line = ""                              # output: build one row as a string
    for _ in range(row):                   # output: add 'row' stars
        line += "*"                        # output: append one star each time
    print(line)                            # output: prints *, **, ***, ****, *****

print()                                    # output: blank line
````

````python
# B) Right-aligned triangle (spaces + stars)
for row in range(1, n + 1):                # output: row = 1..5
    spaces = ""                            # output: build leading spaces
    stars = ""                             # output: build stars
    for _ in range(n - row):               # output: number of spaces on the left
        spaces += " "
    for _ in range(row):                   # output: number of stars on the right
        stars += "*"
    print(spaces + stars)                  # output: right-aligned triangle

print()                                    # output: blank line
````

````python
# C) Solid square (n x n)
for i in range(n):                         # output: repeat n rows
    line = ""                              # output: build one full row
    for i in range(n):                     # output: put n stars in the row
        line += "*"
    print(line)                            # output: prints ***** (n times)

print()                                    # output: blank line
````

````python
# D) Centered pyramid (odd widths: 1,3,5,...)
for row in range(1, n + 1):                # output: row = 1..5
    spaces = ""                            # output: leading spaces to center
    stars  = ""                            # output: odd number of stars
    for _ in range(n - row):               # output: left padding
        spaces += " "
    for _ in range(2 * row - 1):           # output: star count grows by 2 each row
        stars += "*"
    print(spaces + stars)                  # output: centered pyramid row
````

#### ‚ÄúWeekday Vibes Game‚Äù 

- The program asks the user to pick a number 1‚Äì7 (or type q to quit).

- Based on the number, it prints a short, fun quote/vibe using an if‚Äìelif‚Äìelse ladder.

- It keeps running until the user enters q.

<div class="alert alert-block alert-warning">
    <b><font size="5"> Live Exercise Solution</font> </b>
</div>

````python
# Weekday Vibes ‚Äî if/elif ladder + while loop (no packages, no functions)

print("Weekday Vibes üéà  Choose 1‚Äì7 for a vibe, or 'q' to quit.")

while True:                                   # keep asking until the user quits
    s = input("Pick a number (1‚Äì7) or 'q' to quit: ").strip()
    
    if s == "q":                               # quit command
        print("Goodbye! üåü")
        break
    
    if not s in "1234567":                        # simple validation: only digits allowed here
        print("Please enter a number between 1 and 7, or 'q' to quit.")
        continue
    
    n = int(s)                                 # convert the input to an integer
    
    # if‚Äìelif‚Äìelse ladder for 1..7
    if n == 1:
        print("‚ë† Start where you are. Use what you have. Do what you can.")
    elif n == 2:
        print("‚ë° Two is company‚Äîcheck in on a friend today.")
    elif n == 3:
        print("‚ë¢ Three steps: breathe, smile, proceed. üòä")
    elif n == 4:
        print("‚ë£ Four-tune favors the bold‚Äîtry a tiny risk.")
    elif n == 5:
        print("‚ë§ High five! Celebrate a small win right now. üôå")
    elif n == 6:
        print("‚ë• Six-word pep talk: You‚Äôre doing better than you think.")
    elif n == 7:
        print("‚ë¶ Lucky seven! Something good is coming your way. üçÄ")
    else:
        print("That‚Äôs outside 1‚Äì7. Try again!")  # handles numbers not in range
````

### Mini Game ‚Äî High-Low Dice Dash

**Rules:**
- You type **high** (4‚Äì6) or **low** (1‚Äì3).
- The computer **rolls a die**.
- If your call matches the roll‚Äôs category ‚Üí **you score a point**.
- Play up to **5 valid rounds** or type **quit** to stop.

````python
# ============================================
# High-Low Dice Dash (loops + conditionals)
# Each line has:  # output: explanation
# ============================================
import random                                  # output: lets us roll a random die number

print("üé≤ High-Low Dice Dash ‚Äî type 'high', 'low', or 'quit'")  # output: show game title and options

score = 0                                      # output: your points start at 0
rounds = 0                                     # output: counts how many valid rounds you've played
MAX_ROUNDS = 5                                 # output: we'll cap the game at 5 valid rounds

while rounds < MAX_ROUNDS:                     # output: loop until we reach 5 valid rounds
    guess = input("Your call (high/low/quit): ").strip().lower()  # output: read guess, normalize spaces/case
    
    if guess == "quit":                        # output: if user wants to stop
        print("üëã Thanks for playing!")        # output: goodbye message
        break                                  # output: exit the loop immediately
    
    if guess != "high" and guess != "low":     # output: invalid input guard
        print("‚ùó Please type exactly 'high' or 'low'. Try again.")  # output: tell user to retry
        continue                               # output: do NOT count this as a round; restart loop
    
    roll = random.randint(1, 6)                # output: roll a fair die (1..6)
    result = "high" if roll >= 4 else "low"    # output: map the die to 'high' (4‚Äì6) or 'low' (1‚Äì3)
    
    print("üéØ Die shows:", roll)               # output: reveal the die result
    
    if guess == result:                        # output: did the player guess correctly?
        score += 1                             # output: add a point
        print("‚úÖ Nice call! +1 point")        # output: success message
    else:
        print("‚ùå Missed this one!")           # output: miss message
    
    rounds += 1                                # output: we finished a valid round, increment counter
    print("Score:", score, "| Round:", rounds, "/", MAX_ROUNDS)  # output: status line
    print("-" * 30)                            # output: simple separator for readability

# After the loop: show final summary (this runs on quit OR after 5 rounds)
print("\nüèÅ Final Score:", score, "out of", rounds)  # output: final scoreboard

# Tiny celebratory note using a conditional for flavor (optional)
if rounds > 0 and score == rounds:             # output: perfect game check
    print("üåü PERFECT GAME! Legendary calls.")
elif rounds > 0 and score >= (rounds // 2 + rounds % 2):   # output: more wins than losses
    print("üôå Solid instincts!")
elif rounds > 0:
    print("üôÇ Good warm-up‚Äîluck swings both ways!")
````
